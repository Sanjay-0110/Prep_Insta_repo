#### Category Wise Users Distribution

```python
daily_activity_users = daily_activity.groupby('Id').agg({'TotalSteps':'mean', 'Calories':'mean'})
daily_activity_users
```

```python
def categoriser(i):
  if i < 5000:
    return 'Sedentary'
  elif i >= 5000 and i < 7000:
    return 'Lightly Active'
  elif i >= 7000 and i <= 10000:
    return 'Fairly Active'
  else :
    return 'Very Active'

# Add New Column 'Category' & apply above function

daily_activity_users['Category'] = daily_activity_users['TotalSteps'].apply(categoriser)
daily_activity_users.head()
```

```python
activity_category = daily_activity_users['Category'].value_counts()
activity_category

custom_colors = ['Red','gold','lightgreen','aquamarine']

activity_category.plot(kind='pie', figsize = (8,4), labels = None, autopct = '%0.1f%%', fontsize=13, colors=custom_colors, startangle=0, wedgeprops=dict(edgecolor='w'))
plt.title('Users Category Based On Dailysteps', fontdict={'fontweight': 'bold'})


plt.axis('equal')
plt.tight_layout()                                                              #<-- this function adujst sapce and avoid overlaping of details in chart
plt.legend(labels=activity_category.index, prop={'size': 12})
```

Among all users, the distribution based on activity levels is as follows:

* Sedentary: Approximately 24.2% of users take less than 5000 steps per day.
* Lightly Active: Around 15.2% of users take between 5000 and 7000 steps per day.
* Fairly Active: Roughly 39.4% of users take between 7000 and 10,000 steps per day.
* Very Active: About 21.2% of users take more than 10,000 steps per day.

#### User's Total Avg Steps Through Weekdays¶
finding the total average steps activity on weekdays provides valuable insights that can inform various aspects of health, fitness, user engagement.

```python
daily_activity['Day'] = daily_activity['Date'].dt.day_name()
daily_activity
```

```python
daily_activity.groupby('Day').agg({'TotalSteps':'mean'}).sort_values('TotalSteps',ascending=False)
```

```python
from pandas.api.types import CategoricalDtype

ctgs = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# define catergorical datatype for weekdays
ctgs_type = CategoricalDtype(categories=ctgs, ordered=True)

# Assign 'Day' column datatype to 'ctgs_type'
daily_activity['Day'] = daily_activity['Day'].astype(ctgs_type)

# now group data on column 'Day' and by it's mean value, assign subset into new vaiable df 'df_weekday'
df_weekday = daily_activity.copy().groupby('Day').mean().reindex(ctgs)
```

```python
df_weekday.filter(['TotalSteps','TotalDistance','Calories'])
```

```python
weekdays_summary = df_weekday.filter(['TotalSteps','TotalDistance','Calories'])

plt.figure(figsize=(10,3))

sns.set(style="whitegrid")

sns.lineplot(data = weekdays_summary['TotalSteps'], marker='o', markersize=8, color='c')
plt.xlabel("Day of the Week")
plt.ylabel("Total Steps")
plt.title("Total Avg Steps Days Of Week Wise")

plt.grid(True)

plt.xticks(rotation=45)
plt.yticks(fontsize=12)
```

By analyzing the data through graphs, we can discern patterns in user activity throughout the weekdays.

Key Takeways :

Saturday & Tuesday: These are the most active days with average total steps consistently surpassing 8000.

Sunday: Notably less active, with a lower average step count.

Weekday Decline: Following Tuesday, there is a gradual decline in activity levels, indicating reduced physical activity as the week advances.

These insights can help us understand how users' activity patterns vary throughout the weekdays, aiding in the development of strategies to encourage consistent physical activity.

#### App Use Frequancy by User

```python
app_use_frequency = daily_activity.groupby('Id')['Date'].count().reset_index()
app_use_frequency.rename(columns={'Date': 'AppUseFrequency'}, inplace=True)

app_use_frequency.sort_values('AppUseFrequency').head(8)
```

```python
def freq_ctg(num):
    if num >= 1 and num <= 10:
        return "low use"
    elif num >= 11 and num <=20:
        return "moderate use"
    elif num >=21 and num <=31:
        return "high use"
    
app_use_frequency['Category']= app_use_frequency['AppUseFrequency'].apply(freq_ctg)
app_use_frequency
```

By analyzing the App Use Frequency data for each of the 33 users throughout the month, we can discern a notable pattern: a considerable number of users exhibit infrequent app usage. This observation serves as a valuable foundation for implementing tailored strategies aimed at identifying users with limited engagement. These insights enable the application of personalized approaches to effectively enhance user engagement. The outcome is an improved user experience achieved through the identification of less engaged users and the subsequent implementation of strategies to elevate their interaction with the app.

```python
user_freq = app_use_frequency['Category'].value_counts()
user_freq 

colorop = ['#66b3ff', '#99ff99', '#ffcc99']

user_freq.plot(kind='pie',figsize=(8,5), autopct='%0.1f%%', colors = colorop, fontsize = '10', wedgeprops=dict(edgecolor='w'))

plt.title('Daily Use of App', fontdict={'fontweight': 'bold'})

plt.legend()

plt.axis('equal')
plt.tight_layout()

plt.axis('equal')
plt.tight_layout()                                                              #<-- this function adujst sapce and avoid overlaping of details in chart
plt.legend(labels=user_freq.index, prop={'size': 12})
```

When considering the entire user base, the distribution of users based on app usage frequency categories is as follows:
